```python
import pandas as pd
pd.set_option('display.max_columns', None)
import sqlite3
```

```python
# preprocessing/NBA-Boxscore-Database.sqlite
conn = sqlite3.connect('NBA-Boxscore-Database.sqlite')
c = conn.cursor()

# check the table name
c.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(c.fetchall())
```

```python
# check the player_stats
c.execute("SELECT * FROM player_stats")
player_stats = pd.DataFrame(c.fetchall())
player_stats.columns = [x[0] for x in c.description]
player_stats.head()
```

```python
# check the game_info
c.execute("SELECT * FROM game_info")
game_info = pd.DataFrame(c.fetchall())
game_info.columns = [x[0] for x in c.description]
game_info.head()
```

```python
player_stats = player_stats[["game_id", "player", "team", "FG", "FGA", "3P", "3PA", "FT", "FTA", "ORB", "DRB", "AST", "STL", "BLK", "PTS"]]
player_stats.head()
```
