```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# import data
all_seasons = pd.read_csv('all_seasons.csv')
all_seasons.head()
```

```python
# change the season column location to front of player_name
all_seasons = all_seasons[['season', 'player_name'] + [col for col in all_seasons.columns if col not in ['season', 'player_name']]]
# drop Unnamed: 0
all_seasons = all_seasons.drop(columns=['Unnamed: 0'])
all_seasons.head()
```

```python
# new column for filter the season
all_seasons['season_filter'] = all_seasons['season'].str[:4]
# only keep the season after 2004
all_seasons = all_seasons[all_seasons['season_filter'] > '2004']
```

```python
# sort by player_name and season
all_seasons = all_seasons.sort_values(by=['player_name', 'season'], ascending=[True, True])
# reset index
all_seasons = all_seasons.reset_index(drop=True)
all_seasons.head(10)
```

```python
# only keep season, season_filter,player_name, team_abbreviation, age, player_height, player_weight, college, country, draft_year
all_seasons = all_seasons[['season', 'season_filter', 'player_name', 'team_abbreviation', 'age', 'player_height', 'player_weight', 'college', 'country', 'draft_year']]
all_seasons.head()
```

```python
# preprocessing/NBA-Boxscore-Database.sqlite
import sqlite3
conn = sqlite3.connect('NBA-Boxscore-Database.sqlite')
c = conn.cursor()

# check the table name
c.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(c.fetchall())
```

```python
# check the player_stats
c.execute("SELECT * FROM player_stats")
player_stats = pd.DataFrame(c.fetchall())
player_stats.columns = [x[0] for x in c.description]
player_stats.head()
```

```python
# sort by player_name and game_date
player_stats = player_stats.sort_values(by=['player'], ascending=[True])
# reset index
player_stats = player_stats.reset_index(drop=True)
player_stats.head(10)
```

```python
# check game_info
c.execute("SELECT * FROM game_info")
game_stats = pd.DataFrame(c.fetchall())
game_stats.columns = [x[0] for x in c.description]
game_stats.head()
```

```python
# join player_stats and game_info
player_game = pd.merge(player_stats, game_stats, left_on='game_id', right_on='game_id')
player_game.head(10)
```

```python
# add season column
# set August 1st as the start of the season
player_game['season'] = player_game['date'].apply(lambda x: str(int(x[:4]) - 1) if x[5:7] < '08' else x[:4])
player_game.head()
```

```python
# move season column to front
player_game = player_game[['season'] + [col for col in player_game.columns if col not in ['season']]]
player_game.head()
```
