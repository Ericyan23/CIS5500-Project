```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# import data
all_seasons = pd.read_csv('all_seasons.csv')
all_seasons.head()
```

```python
# change the season column location to front of player_name
all_seasons = all_seasons[['season', 'player_name'] + [col for col in all_seasons.columns if col not in ['season', 'player_name']]]
# drop Unnamed: 0
all_seasons = all_seasons.drop(columns=['Unnamed: 0'])
all_seasons.head()
```

```python
# new column for filter the season
all_seasons['season_filter'] = all_seasons['season'].str[:4]
# only keep the season after 2004
all_seasons = all_seasons[all_seasons['season_filter'] > '2004']
```

```python
# sort by player_name and season
all_seasons = all_seasons.sort_values(by=['player_name', 'season'], ascending=[True, True])
# reset index
all_seasons = all_seasons.reset_index(drop=True)
all_seasons.head(10)
```

```python
# only keep season, season_filter,player_name, team_abbreviation, age, player_height, player_weight, college, country, draft_year
all_seasons = all_seasons[['season', 'season_filter', 'player_name', 'team_abbreviation', 'age', 'player_height', 'player_weight', 'college', 'country', 'draft_year']]
all_seasons.head()
```

```python
# split to 3nf
# player
player = all_seasons[['player_name', 'college', 'country', 'draft_year']]
player = player.drop_duplicates()
player = player.reset_index(drop=True)
player.head(10)
```

```python
# seasons info
# Season Stats Table (SeasonStats): This table would store season, player_name, team_abbreviation, age, height, and weight. The player_name would act as a foreign key to link back to the Player table.
season = all_seasons[['season', 'player_name', 'team_abbreviation', 'age', 'player_height', 'player_weight']]
# set the age to integer
season['age'] = season['age'].astype(int)
season = season.drop_duplicates()
# drop the same name
season = season.drop_duplicates(subset=['season', 'player_name'], keep='first')
# reset index
season = season.reset_index(drop=True)
season.head(10)
```

```python
# output to csv
player.to_csv('player.csv', index=False)
season.to_csv('season.csv', index=False)
```
